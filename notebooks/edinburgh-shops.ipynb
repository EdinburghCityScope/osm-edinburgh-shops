## Clustered map of shops

```python
import folium

map = folium.Map(
    location=[55.940167, -3.2394458],
    zoom_start=11
)
markers = folium.MarkerCluster("Shops").add_to(map)

geo_json = folium.GeoJson(open('../data/edinburgh-shops.geojson', encoding='utf8'))
markers.add_child(geo_json)

map
```

## Heatmap of shop locations

```python
import folium
from folium import plugins
import csv

heatmap = folium.Map(
    location=[55.940167, -3.2394458],
    zoom_start=11
)

with open('../data/edinburgh-shops.csv', encoding='utf8') as f:
    reader = csv.reader(f)
    next(reader)
    points = [[row[1], row[2]] for row in reader]

folium.plugins.HeatMap(points, min_opacity=0.35).add_to(heatmap)

heatmap
```
